```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('OTg6QzA_activities.csv')
print(df.head())
print(df.shape)
```

```python
X=df.X.to_numpy()
print(X[0])
Y=df.Y.to_numpy()
print(Y[0])

plt. scatter(X,Y)
plt.show()
```

```python
plt.figure(facecolor='pink')

plt.hist2d(X,Y,bins=30, cmap='spring')

plt.colorbar(label='빈도')
plt.title('동선의 위치값 분석',fontsize=15)
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
plt.figure(figsize=(8,6), facecolor='pink')
plt.hexbin(X,Y,gridsize=20, cmap='spring')

cb=plt.colorbar()
cb.set_label('frequency of appearanxe, gridsize=20')

plt.show()
```

```python
x_quartiles = np.percentile(X, [25, 50, 75])
y_quartiles = np.percentile(Y, [25, 50, 75])

print(f'X 분할 기준: {x_quartiles}')
print(f'Y 분할 기준: {y_quartiles}')

X1, Y1 = X[X <= x_quartiles[0]], Y[X <= x_quartiles[0]]
X2, Y2 = X[(X > x_quartiles[0]) & (X <= x_quartiles[1])], Y[(X > x_quartiles[0]) & (X <= x_quartiles[1])]
X3, Y3 = X[(X > x_quartiles[1]) & (X <= x_quartiles[2])], Y[(X > x_quartiles[1]) & (X <= x_quartiles[2])]
X4, Y4 = X[X > x_quartiles[2]], Y[X > x_quartiles[2]]

# 1등분 그래프
plt.figure(figsize=(6, 6), facecolor='white')
plt.scatter(X1, Y1, color='pink')
plt.title('1st Quartile: X and Y')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.show()

# 2등분 그래프
plt.figure(figsize=(6, 6), facecolor='white')
plt.scatter(X2, Y2, color='lightpink')
plt.title('2nd Quartile: X and Y')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.show()

# 3등분 그래프
plt.figure(figsize=(6, 6), facecolor='white')
plt.scatter(X3, Y3, color='hotpink')
plt.title('3rd Quartile: X and Y')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.show()

# 4등분 그래프
plt.figure(figsize=(6, 6), facecolor='white')
plt.scatter(X4, Y4, color='deeppink')
plt.title('4th Quartile: X and Y')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.show()
```

```python
Z = np.zeros_like(X)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X, Y, Z, color='hotpink')

ax.set_xlabel('X Position')
ax.set_ylabel('Y Position')
ax.set_zlabel('Z Position (Placeholder)')

ax.set_title('3D')

plt.show()
```

```python
x_quartiles = np.percentile(X, 25)

# Q1 구간 제외 (X <= Q1인 값 제외)
mask = X > x_quartiles 
X_filtered = X[mask]
Y_filtered = Y[mask]

plt.figure(figsize=(8, 6))
plt.scatter(X_filtered, Y_filtered, color='hotpink')

plt.xlabel('X Position')
plt.ylabel('Y Position')

plt.show()
```
